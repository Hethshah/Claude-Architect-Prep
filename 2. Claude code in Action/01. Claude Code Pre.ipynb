# Claude Code in Action — How Coding Assistants Work

---

## What a Coding Assistant Actually Does

When given a task like "fix this bug", a coding assistant follows the same
process a human developer would:
```
1. Gather context   → understand the error, find relevant files
2. Formulate a plan → decide how to fix it, what to test
3. Take action      → edit files, run commands, verify the fix
```

Steps 1 and 3 require **interacting with the outside world** — reading files,
running commands, fetching docs. A language model alone cannot do this.

---

## The Tool Use Solution

Language models can only process text and return text. They cannot read files
or run commands by themselves. Tool use solves this with a structured loop:
```
You: "What code is in main.go?"
         |
         v
Coding assistant adds tool instructions to your message
         |
         v
Claude responds: "ReadFile: main.go"
         |
         v
Coding assistant reads the actual file
         |
         v
File contents sent back to Claude
         |
         v
Claude gives a final answer based on the real file contents
```

Claude is generating formatted text that **signals what action to take** —
the coding assistant's code handles the actual execution.

---

## Why Claude's Tool Use is Particularly Strong

Not all language models use tools equally well. Claude (Opus, Sonnet, Haiku)
is specifically strong at understanding what tools do and combining them
effectively for complex tasks.

| Benefit | Detail |
|---|---|
| **Tackles harder tasks** | Combines multiple tools and adapts to tools it hasn't seen before |
| **Extensible platform** | Add new tools to Claude Code — Claude adapts to use them automatically |
| **Better security** | Navigates codebases without requiring full indexing — avoids sending your entire codebase to external servers |

---

## Key Takeaways

- Coding assistants use LLMs to complete programming tasks
- LLMs need tools to handle most real-world coding tasks
- Not all LLMs use tools with the same skill — Claude is particularly capable
- Claude's strong tool use enables better security, customisability, and longevity in Claude Code

# Claude Code in Action — Three Real Examples

---

## The Toolbox

Claude Code's built-in tools and what they do:

| Tool | Role |
|---|---|
| `bash` | Run any shell command |
| `read` / `write` | Read and create files |
| `grep` | Search inside file contents |
| `ls` / `glob` | See what's in a folder |
| `webfetch` | Look things up online |
| `notebookRead` / `notebookEdit` | Work with Jupyter notebook cells |
| `agent` | Delegate a subtask to a sub-agent |
| `todoRead` / `todoWrite` | Track what's left to do |

These are always available. MCP Servers add on top of these for external systems.

---

## Example 1 — GitHub PR Review

**Task:** "Review changes in this pull request"
```
Claude Code gets:
  → Task (review the PR)
  → GitHub MCP Server (tools to interact with GitHub)

Steps Claude takes:
  1. Read the PR changes          ← Read tool
  2. Assess code quality          ← Claude's own reasoning
  3. Write a summary report       ← Write tool
```

**Key insight:** The GitHub MCP Server gives Claude a remote control for GitHub —
list PRs, read diffs, post comments. Without it, Claude cannot touch GitHub at all.

---

## Example 2 — UI Styling (Browser Control)

**Task:** "Improve the design of this app's chat interface and header"
```
Claude Code gets:
  → Task (improve the UI)
  → Playwright MCP Server (tools to control a real browser)

Steps Claude takes:
  1. Open browser                 ← Playwright tool
  2. Navigate to the app          ← Playwright tool
  3. Take a screenshot            ← environment inspection!
  4. Update the styling           ← Edit / Write tool
```

**Key insight:** Step 3 is environment inspection — Claude visually sees the app
before making changes. It is not guessing what the UI looks like. It is looking.

---

## Example 3 — Data Analysis (Jupyter Notebook)

**Task:** "Analyse data in streaming.csv — find insights"
```
No MCP Server needed — built-in tools are sufficient

Steps Claude takes:
  1. Write code in notebook to examine data format  ← NotebookEdit
  2. Run the code, examine output                   ← Bash + NotebookRead
  3. Add more cells, execute each iteratively       ← environment inspection loop
  4. Write a final summary                          ← Write tool
```

**Key insight:** Claude does not write all the code at once and hope for the best.
It runs → observes → adapts → runs again. Every cell execution is a mini
inspection loop.

---

## The Pattern Behind All Three
```
Built-in Tools          MCP Servers
(always available)   +  (plug in as needed)
       |                       |
       +----------+------------+
                  |
            Claude Code
                  |
          +-------+--------+
          |                |
     Do the task     Inspect results
          |                |
          +-------+--------+
                  |
           Loop until done
```

MCP Servers let Claude reach into external systems (GitHub, browsers, databases)
it otherwise could not touch.
