# Claude Code — Local Setup

---

## Installation

Full docs: https://code.claude.com/docs/en/quickstart
```bash
# macOS (Homebrew)
brew install --cask claude-code

# macOS, Linux, WSL
curl -fsSL https://claude.ai/install.sh | bash

# Windows CMD
curl -fsSL https://claude.ai/install.cmd -o install.cmd && install.cmd && del install.cmd
```

After installation, run `claude` in your terminal. First run will prompt you to authenticate.

# Claude Code — Context Management

---

## Why Context Management Matters

> Too much irrelevant context **decreases** Claude's performance.
> Guide Claude toward only the files and docs it actually needs.

---

## The `/init` Command

Run this once when starting Claude in a new project:
```bash
/init
```

Claude will:
- Analyse your entire codebase
- Identify the project's purpose, architecture, key files, and coding patterns
- Write a summary to `CLAUDE.md`

When prompted for permission to write files, either:
- Press `Enter` to approve each write individually
- Press `Shift+Tab` to let Claude write files freely for the session

---

## The `CLAUDE.md` File

Included in **every request** you make — acts like a persistent system prompt
for your project.

**Two purposes:**
- Guides Claude through your codebase (architecture, key commands, coding style)
- Lets you give Claude specific custom instructions

---

## Three `CLAUDE.md` Locations

| File | Shared? | Purpose |
|---|---|---|
| `CLAUDE.md` | Yes — commit to source control | Generated by `/init`, shared with the whole team |
| `CLAUDE.local.md` | No — gitignore it | Personal instructions and customisations |
| `~/.claude/CLAUDE.md` | No — machine-wide | Instructions you want Claude to follow on every project |

---

## Adding Custom Instructions — Memory Mode

Use `#` to enter memory mode and edit `CLAUDE.md` intelligently:
```
# Use comments sparingly. Only comment complex code.
```

Claude merges this instruction into your `CLAUDE.md` automatically.
Use this whenever Claude's behaviour needs adjusting (e.g. too many comments,
wrong code style, incorrect assumptions about the project).

---

## Mentioning Files with `@`

Pull specific files into context on demand:
```
How does the auth system work? @auth
```

Claude shows a list of matching files → you select one → its contents are
included in the request.

Use `@` when you know which files are relevant and want Claude to focus there.

---

## Referencing Files in `CLAUDE.md`

You can use `@` syntax inside `CLAUDE.md` to auto-include files in every request:
```markdown
The database schema is defined in @prisma/schema.prisma.
Reference it anytime you need to understand the structure of data
stored in the database.
```

Claude will include that file's contents automatically on every request —
useful for schemas, config files, or anything frequently referenced.

---

## Context Management Summary

| Tool | When to use |
|---|---|
| `/init` | First time in a new project |
| `CLAUDE.md` | Shared project-wide context and instructions |
| `CLAUDE.local.md` | Personal preferences, not shared with team |
| `~/.claude/CLAUDE.md` | Instructions for all projects on your machine |
| `@file` in chat | Pull a specific file into one request |
| `@file` in `CLAUDE.md` | Auto-include a file in every request |
| `#instruction` | Update `CLAUDE.md` via memory mode |

---

## Key Takeaway

> `CLAUDE.md` is your project's persistent system prompt.  
> Use `/init` to generate it, `#` to update it, and `@` to pull in
> specific files when needed.  
> Good context management = faster, more accurate responses from Claude.

# Claude Code — Screenshots, Planning Mode & Thinking Modes

---

## Using Screenshots

Paste a screenshot directly into Claude to communicate visually:
```
Ctrl+V   ← paste screenshot (not Cmd+V on macOS)
```

Then describe what you want changed. Claude sees exactly what you see —
no need to describe the UI in words.

---

## Planning Mode

For complex tasks that require broad codebase exploration before making changes.

**Enable:**
```
Shift + Tab  (twice from normal mode, once from auto-accept mode)
```

**What Claude does in Planning Mode:**
- Reads more files across the project
- Creates a detailed implementation plan
- Shows exactly what it intends to do
- Waits for your approval before making any changes

Use this when you want to review and redirect Claude before it touches your code.

---

## Thinking Modes

Allow Claude to spend more time reasoning before responding.
Each level gives Claude progressively more tokens for deeper analysis.

| Mode | Reasoning depth |
|---|---|
| `think` | Basic reasoning |
| `think more` | Extended reasoning |
| `think a lot` | Comprehensive reasoning |
| `think longer` | Extended time reasoning |
| `ultrathink` | Maximum reasoning capability |

Just type the phrase in your message — Claude activates the corresponding mode.

---

## Planning vs Thinking — When to Use Each

| | Planning Mode | Thinking Mode |
|---|---|---|
| **Best for** | Broad codebase understanding, multi-file changes | Complex logic, debugging, algorithms |
| **Claude does** | Explores files, creates a plan, waits for approval | Reasons deeply before responding |
| **Activate with** | `Shift + Tab` | Type `think` / `ultrathink` etc. in message |
| **Token cost** | Higher | Higher |

You can **combine both** for tasks that need breadth (planning) and depth (thinking).

---

## Key Takeaway

> Screenshots remove ambiguity — show Claude exactly what to change.  
> Planning Mode gives you a review checkpoint before Claude edits your code.  
> Thinking Modes give Claude more reasoning time for hard problems.  
> Both cost extra tokens — use them when the task genuinely needs it.

# Claude Code — Conversation Control

---

## Interrupting Claude — `Escape`

Press `Escape` to stop Claude mid-response and redirect it.

**When to use:**
- Claude is heading in the wrong direction
- Claude is trying to tackle too many things at once
- You want it to focus on one specific task first

**Example:** Claude starts planning tests for all functions at once →
press `Escape` → "Focus on just the login function first."

---

## Escape + Memory — Fixing Repetitive Errors

When Claude makes the same mistake repeatedly:
```
1. Press Escape     → stop the current response
2. Type #           → open memory mode
3. Add the correct  → e.g. "# Always use async/await, never .then()"
   instruction
4. Continue         → Claude won't repeat the mistake in future sessions
```

---

## Rewinding — `Escape` Twice

Press `Escape` twice to see all messages you've sent and jump back to an
earlier point in the conversation.

**Why this helps:**
- Removes distracting back-and-forth (e.g. a long debugging session)
- Keeps the valuable context Claude has built (codebase understanding)
- Lets you continue from a clean point without losing good history

---

## Context Management Commands

### `/compact`
Summarises the entire conversation while preserving key knowledge Claude has gained.
```
/compact
```

**Use when:**
- Claude has learned a lot about your project
- The conversation is long but contains important context
- You want to continue with related tasks without losing that knowledge

---

### `/clear`
Completely removes conversation history — fresh start.
```
/clear
```

**Use when:**
- Switching to a completely different, unrelated task
- Current context would confuse Claude for the new task
- You want to start over with zero prior context

---

## Quick Reference

| Technique | Shortcut | When to use |
|---|---|---|
| Stop Claude | `Escape` | Wrong direction, too broad |
| Fix recurring errors | `Escape` then `#` | Claude repeats the same mistake |
| Rewind conversation | `Escape` twice | Remove irrelevant history, keep good context |
| Summarise and continue | `/compact` | Long session, related tasks ahead |
| Full reset | `/clear` | Unrelated new task, start fresh |

---

## Key Takeaway

> These are not just convenience features — they are essential tools
> for keeping long AI-assisted sessions focused and productive.  
> Use `Escape` to redirect, `/compact` to preserve knowledge,
> and `/clear` to start clean.

# Claude Code — Custom Commands

---

## What Custom Commands Are

Built-in commands use `/` (e.g. `/init`, `/compact`). You can create your own
to automate repetitive project-specific workflows.

---

## Setup — Folder Structure
```
your-project/
└── .claude/
    └── commands/
        ├── audit.md          → creates /audit command
        └── write_tests.md    → creates /write_tests command
```

The **filename** becomes the command name.
After creating a command file, **restart Claude Code** for it to be recognised.

---

## Example 1 — `/audit` (No Arguments)

File: `.claude/commands/audit.md`
```markdown
Run a security audit on the project dependencies:

1. Run `npm audit` to identify vulnerable packages
2. Run `npm audit fix` to apply available updates
3. Run the test suite to verify nothing was broken by the updates
```

Usage:
```
/audit
```

---

## Example 2 — `/write_tests` (With Arguments)

File: `.claude/commands/write_tests.md`
```markdown
Write comprehensive tests for: $ARGUMENTS

Testing conventions:
* Use Vitest with React Testing Library
* Place test files in a __tests__ directory in the same folder as the source file
* Name test files as [filename].test.ts(x)
* Use @/ prefix for imports

Coverage:
* Test happy paths
* Test edge cases
* Test error states
```

Usage:
```
/write_tests the use-auth.ts file in the hooks directory
```

`$ARGUMENTS` is replaced with whatever you type after the command name.
Arguments can be file paths, descriptions, feature names — any string.

---

## Key Benefits

| Benefit | Detail |
|---|---|
| **Automation** | Turn multi-step workflows into a single command |
| **Consistency** | Same steps followed every time, no steps skipped |
| **Context** | Embed your team's conventions directly in the command |
| **Flexibility** | `$ARGUMENTS` makes commands reusable across different inputs |

---

## Common Use Cases

- Security audits (`/audit`)
- Test generation with project conventions (`/write_tests`)
- Boilerplate generation following team patterns (`/new_component`)
- Deployment checklists (`/deploy`)
- Code review checklists (`/review`)

---

## Key Takeaway

> Custom commands turn your most-repeated Claude Code workflows into
> one-word shortcuts with your project's conventions baked in.  
> Use `$ARGUMENTS` to make them flexible enough to reuse across different inputs.

# Claude Code — Adding MCP Servers

---

## What MCP Servers Add

MCP servers run locally or remotely and give Claude new tools it wouldn't
normally have — browser control, database access, cloud integrations, and more.

---

## Installing the Playwright MCP Server

Run this in your terminal (not inside Claude Code):
```bash
claude mcp add playwright npx @playwright/mcp@latest
```

This registers the server as "playwright" and provides the command to start it.

---

## Managing Permissions

By default Claude asks for permission each time it uses an MCP tool.
To pre-approve a server, edit `.claude/settings.local.json`:
```json
{
  "permissions": {
    "allow": ["mcp__playwright"],
    "deny": []
  }
}
```

Note the **double underscore** in `mcp__playwright`.

---

## Practical Example — Improving Component Generation

With Playwright connected, you can ask Claude to:
```
"Navigate to localhost:3000, generate a basic component, review the
styling, and update the generation prompt at @src/lib/prompts/generation.tsx
to produce better components going forward."
```

**What Claude does:**
```
1. Open browser → navigate to localhost:3000
2. Generate a test component
3. Visually analyse the styling and code quality    ← sees actual output
4. Update the generation prompt file                ← makes informed changes
5. Test the improved prompt with a new component    ← verifies the improvement
```

**Key advantage:** Claude sees the actual visual output, not just the code —
leading to much more informed styling decisions.

**Before:** Generic purple-to-blue gradients, standard Tailwind patterns
**After:** Warm sunset gradients, ocean depth themes, asymmetric layouts,
creative spacing

---

## Other MCP Servers Worth Exploring

| Category | What it enables |
|---|---|
| Database | Query and manage databases directly |
| API testing | Test and monitor APIs |
| File system | Extended file operations |
| Cloud services | AWS, GCP, Azure integrations |
| Dev tool automation | CI/CD, build tools, package managers |

---

## Key Takeaway

> MCP servers transform Claude Code from a code assistant into a
> comprehensive development partner that can interact with your
> entire toolchain.  
> Playwright is the most popular starting point — it gives Claude
> eyes on your actual UI, enabling visual inspection and informed improvements.

# Claude Code — GitHub Integration

---

## What It Does

Runs Claude inside GitHub Actions — enabling two automated workflows:
- **Mention support** — tag `@claude` in any issue or PR
- **Automatic PR reviews** — Claude reviews every pull request automatically

---

## Setup

Run inside Claude Code:
```bash
/install-github-app
```

This walks you through:
1. Installing the Claude Code app on GitHub
2. Adding your API key
3. Auto-generating a PR with the workflow files

Once merged, workflow files appear in `.github/workflows/`.

---

## Default Workflows

### Mention Action
Tag `@claude` in any issue or pull request:
```
@claude add input validation to the login form
```
Claude will:
- Analyse the request and create a task plan
- Execute the task with full codebase access
- Reply with results directly in the issue or PR

### PR Review Action
Triggers automatically on every new pull request:
- Reviews the proposed changes
- Analyses the impact of modifications
- Posts a detailed report on the PR

---

## Customising the Workflow

After merging the initial PR, edit the workflow YAML files to suit your project.

### Add Project Setup Steps
```yaml
- name: Project Setup
  run: |
    npm run setup
    npm run dev:daemon
```

### Add Custom Instructions
```yaml
custom_instructions: |
  The project is already set up with all dependencies installed.
  The server is running at localhost:3000.
  Logs are written to logs.txt.
  Query the DB with the sqlite3 CLI if needed.
  Use mcp__playwright tools to launch a browser and interact with the app.
```

### Add MCP Server Config
```yaml
mcp_config: |
  {
    "mcpServers": {
      "playwright": {
        "command": "npx",
        "args": [
          "@playwright/mcp@latest",
          "--allowed-origins",
          "localhost:3000;cdn.tailwindcss.com;esm.sh"
        ]
      }
    }
  }
```

### Allowed Tools — Must Be Explicit

Unlike local development, GitHub Actions requires every tool to be listed individually:
```yaml
allowed_tools: "Bash(npm:*),Bash(sqlite3:*),mcp__playwright__browser_snapshot,mcp__playwright__browser_click,..."
```

> There is no permission shortcut in GitHub Actions. List every tool you need.

---

## Local vs GitHub Actions — Key Difference

| | Local Claude Code | GitHub Actions |
|---|---|---|
| Permissions | Can pre-approve with wildcard | Must list every tool explicitly |
| Trigger | You type in terminal | `@claude` mention or PR event |
| Setup | Interactive | Defined in YAML workflow file |

---

## Best Practices

- Start with default workflows, customise gradually
- Use `custom_instructions` to give Claude project-specific context
- Be explicit about every tool permission when using MCP servers
- Test with simple tasks before complex ones

---

## Key Takeaway

> The GitHub integration turns Claude into an automated team member —
> reviewing PRs, handling issues, and executing tasks directly in your
> GitHub workflow.  
> The key setup difference from local use: every tool must be explicitly
> permitted in the workflow YAML.
