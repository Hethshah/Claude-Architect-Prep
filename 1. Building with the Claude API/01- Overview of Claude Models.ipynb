# Claude Model Family

## Overview

Anthropic offers three tiers of Claude models, each optimized for different use cases:

| Property | Claude Opus | Claude Sonnet | Claude Haiku |
|---|---|---|---|
| **Description** | Highest level of intelligence | Balances quality, speed, and cost | Most cost-efficient and latency-optimized |
| **Cost** | High | Medium | Low |
| **Comparative Latency** | Moderate | Fast | Fastest |
| **Supports Reasoning** | Yes | Yes | No |

---

## Best Used For

### Claude Opus
- Advanced software development, especially large-scale architecting
- Long running tasks that require sustained focus
- Strategic planning with multi-step problem solving
- Tasks that benefit from advanced reasoning

### Claude Sonnet
- Common coding tasks
- Document creation and editing
- Content marketing and copywriting
- Data analysis and visualization projects
- Image analysis
- Process automation

### Claude Haiku
- Quick code completions and suggestions
- Content moderation and filtering
- Data extraction and categorization
- Language translation
- Q&A systems and knowledge retrieval
- High-volume, straightforward text processing tasks

---

## Key Takeaway

> Choose **Opus** for complex, long-horizon tasks. Choose **Sonnet** for a balanced everyday workhorse. Choose **Haiku** for speed, cost efficiency, and high-volume pipelines.

# Picking the Right Model

There is a fundamental trade-off axis when selecting a Claude model:
```
Intelligence <---------------------------------> Cost / Speed
   Claude Opus       Claude Sonnet        Claude Haiku
```

---

### The Trade-off Summarized

| Model | Intelligence | Cost | Speed |
|---|---|---|---|
| **Claude Opus** | Highest | High | Lower (higher latency) |
| **Claude Sonnet** | Strong | Medium | Fast |
| **Claude Haiku** | Moderate | Low | Fastest |

---

### Decision Rule

- If your task demands **maximum reasoning and accuracy** — use **Opus**, and accept the higher cost and latency.
- If you need a **reliable all-rounder** for most production tasks — use **Sonnet**.
- If you are optimizing for **throughput, speed, or cost at scale** — use **Haiku**.

# How AI Models Perform Reasoning

AI models reason through a combination of learned statistical patterns, architectural mechanisms,
and inference-time strategies.

---

## 1. The Foundation: Transformer Architecture

- Modern LLMs are built on the **Transformer architecture**
- The key component is **self-attention** — lets the model weigh relationships between all tokens
  (words/subwords) in the input simultaneously
- The model does not read left-to-right; it computes a **rich web of relationships across the
  entire context**, deciding which tokens are most relevant to each other

---

## 2. Pattern Matching vs. Structured Reasoning

| Type | Mechanism | Example |
|---|---|---|
| **Associative Recall** | Direct pattern from training | "Paris is the capital of..." |
| **Multi-step Reasoning** | Chaining intermediate steps | Math proofs, logic puzzles |
| **Analogical Reasoning** | Mapping structure across domains | "How is X like Y?" |
| **Causal Reasoning** | Learned if-then relationships | Debugging, root cause analysis |

---

## 3. Chain-of-Thought (CoT) — The Key Reasoning Mechanism

The biggest leap in LLM reasoning came from **Chain-of-Thought prompting and training**.
```
Problem → Step 1 → Step 2 → Step 3 → Answer
```

Why this works:
- Each generated token becomes **new context** that conditions the next token
- The model uses its **own output as working memory**
- Errors can be caught and corrected mid-stream

---

## 4. The KV Cache as "Working Memory"

- As the model generates tokens, past computations are stored in a **Key-Value (KV) cache**
- Acts like **short-term working memory** — the model can look back at prior reasoning steps
  without recomputing them

---

## 5. Inference-Time Compute Scaling

Newer models (o1, o3, DeepSeek-R1, Claude extended thinking) use **extended inference** —
spending more compute at reasoning time:

- The model generates a long internal **scratchpad (thinking tokens)**
- It can explore multiple solution paths, backtrack, and self-correct
- This trades **latency for accuracy** on hard problems

---

## 6. Emergent Capabilities

Reasoning abilities are **not explicitly programmed** — they emerge from scale:

- More parameters → richer internal representations
- More training data → more reasoning patterns absorbed
- **RLHF / RLAIF** fine-tuning → model learns to prefer logically consistent outputs

---

## 7. Known Failure Modes

| Failure | Explanation |
|---|---|
| **No true world model** | Reasoning is over token distributions, not symbolic logic |
| **Context length limits** | Very long reasoning chains can degrade |
| **Hallucination** | Confident but wrong intermediate steps that cascade |
| **Mathematical precision** | Arithmetic errors from treating numbers as tokens |

---

## Key Takeaway

> AI reasoning is fundamentally **next-token prediction**, but chain-of-thought transforms this
> into a powerful iterative process — where the model's own intermediate outputs become the
> scaffolding for progressively deeper conclusions. It mimics, but does not replicate, human
> deliberate thought.

# Transformer Architecture & Self-Attention

---

## Core Idea

> A Transformer takes a sequence of words and figures out which words should "pay attention"
> to which other words — and by how much — before deciding what each word means in context.

---

## Self-Attention: The Intuition

Words do not have fixed meanings in isolation. Self-attention lets every word look at every
other word simultaneously to resolve ambiguity.

**Example — the word "bank":**
```
"I deposited money at the bank"   →  attends to "money", "deposited"  →  financial bank
"I sat by the river bank"         →  attends to "river", "sat"         →  riverbank
```

The same word, different context, different meaning — resolved by attention weights.

---

## Inside the Transformer: Layer by Layer

### 1. Input Tokens + Positional Encoding
- Words are converted to numerical vectors (embeddings)
- Since Transformers process everything **in parallel** (not left-to-right), **positional
  encodings** are injected so the model knows word order

### 2. Self-Attention — The Core Step
Every word asks three questions about every other word:

| Symbol | Name | Question it answers |
|---|---|---|
| **Q** | Query | "What am I looking for?" |
| **K** | Key | "What do I contain / offer?" |
| **V** | Value | "What do I actually contribute if selected?" |

The output is a **weighted blend** of all words' Values, where the weights are determined
by how well each Query matches each Key.

### 3. Add & Normalise (Skip Connection)
- The attention output is **added back to the original input**
- This is a residual / skip connection — prevents information loss and keeps training stable

### 4. Feed-Forward Network
- After words have "consulted" each other via attention, each word is processed
  **independently** through a small neural network
- This is where deeper reasoning happens on each enriched representation

### 5. Repeat N Times (Stacked Layers)
- The full block stacks up — up to **96 layers** in large models
- **Early layers** — simple patterns (grammar, nearby word relationships)
- **Deeper layers** — complex meaning (coreference, reasoning, world knowledge)

---

## Full Data Flow
```
Raw Text
   |
   v
[Token Embeddings + Positional Encoding]
   |
   v
[Self-Attention]  <---+
   |                  |  (skip connection)
   v                  |
[Add & Normalise] ----+
   |
   v
[Feed-Forward Network]  <---+
   |                        |  (skip connection)
   v                        |
[Add & Normalise] ----------+
   |
   v
(Repeat N times)
   |
   v
Output Representation (context-aware meaning of every token)
```

---

## Key Takeaway

> A Transformer is a machine that, for every word, asks **"which other words matter most
> for understanding me?"** — and does this millions of times across many layers, until the
> meaning of every word is fully informed by its entire context.
