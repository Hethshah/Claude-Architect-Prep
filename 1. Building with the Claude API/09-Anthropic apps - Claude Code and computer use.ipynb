---

## Overview

Two Anthropic-built applications that serve as real-world examples of AI agents
in action. Studying them builds intuition for building your own agents.
```
Claude Code     →  Agentic coding assistant in your terminal
Computer Use    →  Tools that let Claude interact with a full desktop environment
```

---

## Claude Code

A terminal-based coding assistant — Claude available directly in your command line.

**What it can do:**
- Edit files and fix bugs
- Answer coding questions
- Help with development workflows
- Execute multi-step programming tasks autonomously

---

## Computer Use

A collection of tools that give Claude access to a full desktop environment.

**What it can do:**
- Browse the internet
- Interact with desktop applications
- Navigate visual interfaces
- Perform tasks that require seeing and clicking — not just text

Both applications demonstrate the same core agent principles:

| Principle | How it shows up |
|---|---|
| **Tool integration** | Claude Code uses file/shell tools; Computer Use uses screen/mouse/keyboard tools |

| **Environmental interaction** | Both perceive and act on a real environment (filesystem, desktop) |
| **Autonomous problem-solving** | Both decide what to do next without step-by-step human instruction |

# Claude Code

---

## What It Is

A terminal-based coding assistant — Claude running directly in your command line,
ready to help with any development task without leaving your terminal.

---

## Capabilities

| Capability | Detail |
|---|---|
| **File operations** | Search, read, and edit files in your project |
| **Terminal access** | Run shell commands directly from the conversation |
| **Web access** | Search docs, fetch code examples, look up APIs |
| **MCP Server support** | Extend with additional tools — databases, APIs, custom services |

The MCP integration means Claude Code is extensible — connect any MCP server
to add specialised tooling for your specific workflow.

---

## Platform Support

Works on **MacOS**, **Windows WSL**, and **Linux**.

---

## Installation — 3 Steps
```bash
# Step 1 — Check if Node.js is installed (install from nodejs.org if not)
npm help

# Step 2 — Install Claude Code globally
npm install -g @anthropic-ai/claude-code

# Step 3 — Start Claude Code and log in
claude
```

On first run, `claude` prompts you to log in to your Anthropic account.

Full setup guide: https://docs.anthropic.com

# Claude Code — MCP Integration

---

## How It Works

Claude Code has a built-in MCP client. Connect any MCP server and Claude
immediately gains access to its tools, resources, and prompts.
```
Claude Code (MCP Client)
        |
        +---> Document MCP Server   → read PDFs, Word docs
        +---> Sentry MCP Server     → fetch and fix bugs
        +---> Jira MCP Server       → read ticket requirements
        +---> Slack MCP Server      → notify team when done
        +---> Any custom MCP server → your internal tools
```

---

## Adding an MCP Server
```bash
claude mcp add [server-name] [command-to-start-server]

# Example — register a local document processing server
claude mcp add documents uv run main.py
```

Once registered, Claude Code connects to the server automatically on startup.

---

## Example — Document Processing Tool

Build an MCP server with a `document_path_to_markdown` tool, then ask:
```
> Convert the tests/fixtures/mcp_docs.docx file to markdown
```

Claude detects the file, calls your custom tool, and returns the converted content.
No extra instructions needed — Claude decides to use the tool automatically.

---

## Popular MCP Integrations

| MCP Server | What it adds to Claude Code |
|---|---|
| `sentry-mcp` | Discover and fix bugs logged in Sentry |
| `playwright-mcp` | Browser automation for testing and debugging |
| `figma-context-mcp` | Access Figma designs directly |
| `mcp-atlassian` | Read Confluence docs and Jira tickets |
| `firecrawl-mcp-server` | Web scraping capabilities |
| `slack-mcp` | Post messages and reply to threads |

---

## Building a Full Development Workflow

Combine multiple servers to match your entire development process:
```
1. Sentry MCP    → fetch production error details
2. Jira MCP      → read the ticket requirements
3. Claude Code   → writes the fix using your codebase
4. Slack MCP     → notifies the team when work is complete
```

Claude handles the entire workflow autonomously — you just describe the task.
